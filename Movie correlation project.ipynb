```python
#Import Libraries
import pandas as pd 
import seaborn as sns 
import numpy as np


import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\adity\Desktop\PythonJupiter Projects\movies.csv\movies.csv')
```

```python
# Let's Look at the data 
df.head()
```

```python
# lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} = {}%'.format(col, pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
# change data type of columns
# First fill NaN values with 0 (or another appropriate value) before converting to int
df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')
df['votes'] = df['votes'].fillna(0).astype('int64')
```

```python
df
```

```python
# create correct year column
df['yearcorrect'] = df['released'].str.split(',').str[1].str[:5].str.strip()
```

```python
df = df.sort_values(by=['score'], inplace=False , ascending=False)
```
